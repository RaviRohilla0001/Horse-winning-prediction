```python
cleaned_DF=pd.DataFrame(cleaned_DF)
```

```python
cleaned_DF.head(5)
```

```python
features=["Distance","Direction","Surface","Condition","Weight Carried","FF (Overall)","Entrants","Horses_name"]
categori=["Direction","Surface","Condition","Horses_name"]
numerical=[f for f in features if f not in categori]


process=ColumnTransformer(
    transformers=[("num",StandardScaler(),numerical),
                  ("cat",OneHotEncoder(),categori)]
    )
X=process.fit_transform(cleaned_DF)
```

```python
with open("process.pkl","wb")as file:
    pickle.dump(process,file)
```

```python
y_position=cleaned_DF["Finish Position"]
y_time=cleaned_DF["Finish Time"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y_time,test_size=0.3)
```

```python
rf=RandomForestRegressor(n_estimators=150,
    max_depth=6,
    n_jobs=-1,
    random_state=42)
rf.fit(x_train,y_train)
```

```python
rf_re_train=rf.predict(x_train)
print(mean_absolute_error(y_train,rf_re_train))
print(r2_score(y_train,rf_re_train))
```

```python
rf_re_test=rf.predict(x_test)
print(mean_absolute_error(y_test,rf_re_test))
print(r2_score(y_test,rf_re_test))
```

```python
xgb=XGBRegressor()
xgb.fit(x_train,y_train)
```

```python
xgb_re_train=rf.predict(x_train)
print(mean_absolute_error(y_train,xgb_re_train))
print(r2_score(y_train,xgb_re_train))
```

```python
xgb_re_test=rf.predict(x_test)
print(mean_absolute_error(y_test,xgb_re_test))
print(r2_score(y_test,xgb_re_test))
```

```python
lgbm=XGBRegressor(n_estimators=200,
    max_depth=5,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    tree_method="hist",
    n_jobs=-1,
    random_state=42)
lgbm.fit(x_train,y_train)
with open("1LGBM.pkl","wb")as file:
    pickle.dump(lgbm,file)
```

```python
lgbm_re_train=lgbm.predict(x_train)
print(mean_absolute_error(y_train,lgbm_re_train))
print(r2_score(y_train,lgbm_re_train))
```

```python
lgbm_re_test=lgbm.predict(x_test)
print(mean_absolute_error(y_test,lgbm_re_test))
print(r2_score(y_test,lgbm_re_test))
```

```python
rf_scores = cross_val_score(rf, X, y_time, cv=5, scoring="r2")
xgb_scores = cross_val_score(xgb, X, y_time, cv=5, scoring="r2")
    
print(rf_scores.mean(), xgb_scores.mean())
```
